```python
import cv2
from matplotlib import pyplot as plt
```

![openCV](./images/opencv.png)


# Reading the image

```python
img = cv2.imread('./images/opencv.png')
name = 'OpenCV'
plt.title(name)
plt.imshow(img)
```

# Flipping the image horizontally

```python
horizontal_flip = cv2.flip(img, 1)
name = 'OpenCV Horizontal Flip'
plt.title(name)
plt.imshow(horizontal_flip)
```

# Flipping the image vertically

```python
vertical_flip = cv2.flip(img, 0)
name = 'OpenCV Vertical Flip'
plt.title(name)
plt.imshow(vertical_flip)
```

# Flipping the iamge vertically and horizontally

```python
vertical_horizontal_flip = cv2.flip(img, -1)
name = 'OpenCV Vertical and Horizontal Flip'
plt.title(name)
plt.imshow(vertical_horizontal_flip)
```
