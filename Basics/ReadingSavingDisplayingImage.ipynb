# Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Reading the image

```python
import cv2
image = cv2.imread('../images/Midjourney.png')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.title('Original Midjourney Image')
plt.imshow(image_rgb)
```

# Saving the grayscale image

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

cv2.imwrite('../images/MidjourneyGray.png', gray_image)

plt.title('GrayScale Midjourney Image')
plt.imshow(gray_image)
```

# Loading and displaying saved image

```python
#Loading
image = cv2.imread('../images/MidjourneyGray.png')

#Displaying
plt.title('Saved Midjourney Image')
plt.imshow(image)
```
